## 1.1 Les Listes en Python

### Qu'est-ce qu'une liste ?

Une **liste** est une collection ordonnée de données. En génie des procédés, on les utilise pour stocker :
- Des mesures expérimentales (température, pression, débit)
- Des compositions de mélanges
- Des résultats de calculs

### Création de listes

````python
# Températures mesurées dans un réacteur (°C)
temperatures = [25, 50, 75, 100, 125]
print("Températures :", temperatures)

# Pressions mesurées (bar)
pressions = [1.0, 1.5, 2.0, 2.5, 3.0]
print("Pressions :", pressions)

# Compositions molaires
compositions = [0.25, 0.35, 0.20, 0.15, 0.05]
print("Compositions :", compositions)
````

### Accéder aux éléments d'une liste

Les indices commencent à **0** en Python !

````python
mesures = [25.5, 26.1, 25.8, 26.3, 25.9]

# Premier élément (indice 0)
print("Première mesure :", mesures[0])

# Deuxième élément (indice 1)
print("Deuxième mesure :", mesures[1])

# Dernier élément (indice -1)
print("Dernière mesure :", mesures[-1])

# Avant-dernier élément (indice -2)
print("Avant-dernière mesure :", mesures[-2])
````

### Opérations de base sur les listes

````python
mesures = [25.5, 26.1, 25.8, 26.3, 25.9]

# Longueur de la liste (nombre d'éléments)
print("Nombre de mesures :", len(mesures))

# Somme des éléments
print("Somme :", sum(mesures))

# Maximum
print("Maximum :", max(mesures))

# Minimum
print("Minimum :", min(mesures))

# Moyenne (somme / nombre)
moyenne = sum(mesures) / len(mesures)
print(f"Moyenne : {moyenne:.2f}")
````

### Modifier une liste

````python
temperatures = [25, 50, 75]
print("Liste initiale :", temperatures)

# Ajouter un élément à la fin
temperatures.append(100)
print("Après append(100) :", temperatures)

# Insérer un élément à la position 1
temperatures.insert(1, 37.5)
print("Après insert(1, 37.5) :", temperatures)

# Supprimer un élément par valeur
temperatures.remove(37.5)
print("Après remove(37.5) :", temperatures)

# Supprimer le dernier élément
temperatures.pop()
print("Après pop() :", temperatures)

# Trier la liste
temperatures.sort()
print("Après sort() :", temperatures)
````

## 1.2 Le Slicing (Découpage)

Le **slicing** permet d'extraire une partie d'une liste.

**Syntaxe** : `liste[debut:fin:pas]`

- `debut` : indice de départ (inclus)
- `fin` : indice de fin (exclu)
- `pas` : intervalle entre les éléments (optionnel)

````python
temperatures = [20, 30, 40, 50, 60, 70, 80, 90, 100]
print("Liste complète :", temperatures)

# Les 3 premiers éléments
print("3 premiers :", temperatures[:3])

# Du 3ème au 6ème élément
print("Indices 2 à 5 :", temperatures[2:6])

# Les 3 derniers éléments
print("3 derniers :", temperatures[-3:])

# Un élément sur deux
print("Un sur deux :", temperatures[::2])

# Liste inversée
print("Inversée :", temperatures[::-1])
````

## 1.3 Les Boucles `for`

La boucle `for` permet de **parcourir** une liste et d'effectuer des opérations sur chaque élément.

### Syntaxe de base

````python
temperatures = [25, 50, 75, 100]

# Parcourir la liste
for T in temperatures:
    print(f"Température : {T}°C")
````

### Utilisation de `range()`

`range(n)` génère une séquence de 0 à n-1

````python
# Afficher les nombres de 0 à 4
for i in range(5):
    print(i)

print()

# De 1 à 10
for i in range(1, 11):
    print(i, end=' ')

print("\n")

# De 0 à 10 par pas de 2
for i in range(0, 11, 2):
    print(i, end=' ')
````

### Utilisation de `enumerate()`

`enumerate()` donne à la fois l'**indice** et la **valeur**

````python
temperatures = [25, 50, 75, 100]

# Avec enumerate (démarre à 0)
for index, T in enumerate(temperatures):
    print(f"Mesure {index} : {T}°C")

print()

# Commencer à 1
for index, T in enumerate(temperatures, 1):
    print(f"Mesure {index} : {T}°C")
````

### Utilisation de `zip()`

`zip()` permet de parcourir **plusieurs listes en même temps**

````python
temperatures = [25, 50, 75, 100]
pressions = [1.0, 1.5, 2.0, 2.5]

# Parcourir les deux listes ensemble
for T, P in zip(temperatures, pressions):
    print(f"T = {T}°C, P = {P} bar")
````

## 1.4 Les Boucles `while`

La boucle `while` continue tant qu'une **condition** est vraie.

**⚠️ Attention** : Il faut s'assurer que la condition devient fausse, sinon boucle infinie !

````python
# Exemple : Montée en pression
pression = 1.0  # bar
temps = 0       # minutes

print("Montée en pression :")
while pression < 5.0:
    print(f"t = {temps} min, P = {pression} bar")
    pression += 0.5  # Augmentation de 0.5 bar
    temps += 1       # Chaque minute

print(f"Pression cible atteinte : {pression} bar")
````

## 1.5 Application : Conversion d'unités

Exemple pratique : Convertir des températures de °C en Kelvin

````python
# Températures en Celsius
T_celsius = [25, 50, 75, 100, 125]

# Méthode 1 : Avec une boucle for
T_kelvin = []
for T in T_celsius:
    T_K = T + 273.15
    T_kelvin.append(T_K)

print("Conversions °C → K :")
for celsius, kelvin in zip(T_celsius, T_kelvin):
    print(f"{celsius}°C = {kelvin} K")

print()

# Méthode 2 : List comprehension (plus concis)
T_kelvin_v2 = [T + 273.15 for T in T_celsius]
print("Méthode 2 :", T_kelvin_v2)
````

### 2.1 Analyse de phases d'une réaction

Dans une réaction, on distingue 3 phases :
- Phase de chauffe : 3 premières mesures
- Phase de réaction : mesures 4 à 8 (indices 3 à 8)
- Phase de stabilisation : 3 dernières mesures

Liste complète : `[25, 35, 48, 65, 82, 95, 105, 110, 112, 113, 113.5, 114]`

**À faire :**
1. Utilisez le slicing pour extraire chaque phase
2. Calculez la température moyenne de chaque phase

````python
# Écrivez votre code ici
temperatures = [25, 35, 48, 65, 82, 95, 105, 110, 112, 113, 113.5, 114]

# Phase de chauffe
phase_chauffe =temperatures[0:3]
print("Phase de chauffe :",phase_chauffe)

# Phase de réaction
phase_reaction =temperatures[3:8]
print("Phase de réaction :",phase_reaction)
# Phase de stabilisation
phase_stabilisation =temperatures[-3:]
print("Phase de stabilisation :",phase_stabilisation)
# Calcul de la température moyenne de chaque phase
moyenne_chauffe = sum(phase_chauffe) / len(phase_chauffe)
moyenne_reaction = sum(phase_reaction) / len(phase_reaction)
moyenne_stabilisation = sum(phase_stabilisation) / len(phase_stabilisation)

# Moyennes
print("Température moyenne de la phase de chauffe :",moyenne_chauffe,"°C")
print("Température moyenne de la phase de réaction :",moyenne_reaction,"°C")
print("Température moyenne de la phase de stabilisation :",moyenne_stabilisation,"°C")
````

## Exercice 4 : Boucles `while`

### 4.1 Montée en pression dans un autoclave

Conditions :
- Pression initiale : 1.0 bar
- Augmentation : 0.5 bar/min
- Pression cible : 10.0 bar

**À faire :** Utilisez une boucle `while` pour simuler la montée en pression

````python
# les conditions
pression = 1.0
temps = 0
P_cible = 10.0
augmentation = 0.5
# boucle while pour simuler la montee en pression
while pression < P_cible:
  print("minute:",temps,"pression:",pression,"bar")
  pression+= augmentation
  temps += 1
  #affichage la montee en pression
  print("minute:",temps,"pression:",pression,"bar")
  print("Pression cible atteinte")
````

### 4.2 Convergence vers l'équilibre

Une concentration évolue selon : $C_{n+1} = 0.9 \times C_n + 0.5$

Conditions :
- Concentration initiale : C = 10.0 mol/L
- Critère d'arrêt : $|C_{n+1} - C_n| < 0.01$

**À faire :** Simulez l'évolution jusqu'à l'équilibre

````python
#  les donnees
C = 10.0
iteration = 0
tolerance = 0.01
#levolution jusqu'a l'equilibre
while True:
  C_nov = 0.9 * C + 0.5
  erreur = abs(C_nov - C)
  if erreur < tolerance:
    break
  C = C_nov
  iteration += 1
#affichage de resultats
print("Concentration finale :",C,"mol/L")
print("Nombre d'itérations :",iteration)
print("Convergence atteinte")
````

## Exercice 5 : Bilan de Matière

### 5.1 Composition d'un mélange

Fractions massiques : `[0.25, 0.30, 0.20, 0.15, 0.10]`

Composants : `['A', 'B', 'C', 'D', 'E']`

**À faire :**
1. Vérifiez que la somme des fractions = 1.0
2. Pour 100 kg de mélange, calculez la masse de chaque composant

````python
# les donnees
fractions = [0.25, 0.30, 0.20, 0.15, 0.10]
composants = ['A', 'B', 'C', 'D', 'E']
masse_totale = 100  # kg
# la somme des fractions
somme_fractions = sum(fractions)
# verification de la somme des fractions
if somme_fractions == 1.0:
  print("La somme des fractions est égale à 1.0")
else:
  print("La somme des fractions n'est pas égale à 1.0")
# calcul des masses pour 100kg
masses = []
for fraction in range(len(fractions)):
  masse = fraction * masse_totale
  masses.append(masse)
  print("La masse de",composants,"est de",masse,"kg")
````

### 5.2 Rendement d'une réaction

Données :
- Masses expérimentales (g) : `[45, 48, 50, 47, 49]`
- Masse théorique : 50 g

**À faire :**
1. Calculez le rendement de chaque expérience : $\eta = \frac{m_{exp}}{m_{theo}} \times 100$
2. Calculez le rendement moyen

````python
# les donnees
masses_exp = [45, 48, 50, 47, 49]
masse_theorique = 50
# calcul des rendements
rendements = []
for masse in masses_exp:
  rendement = (masse/masse_theorique)*100
  rendements.append(rendement)
  print("Le rendement de",rendement,"%")
#calcul du rendement moyen
rendement_moyen = sum(rendements)/len(rendements)
print("Le rendement moyen est de",rendement_moyen,"%")
````

### 5.3 Filtrage : Températures hors norme

Températures mesurées : `[85, 92, 98, 105, 110, 95, 88, 115, 90, 108]`

````python
# les temperatures
temperatures = [85, 92, 98, 105, 110, 95, 88, 115, 90, 108]
# les temperatures > 100
surchauffes = []
for temperature in temperatures:
  if temperature > 100:
    surchauffes.append(temperature)
# affichage de resultats
print("Les temperatures > 100°C sont :",surchauffes)
````
